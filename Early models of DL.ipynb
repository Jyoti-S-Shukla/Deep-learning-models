```python
## Part A : MP model

import numpy as np
import matplotlib.pyplot as plt

#Output=1
def mp_model(x,gate):
 
  ## Write the activation function and define weights and bias for all three gates
  Output = 0
  if gate == "NAND":

    ## Write your code here
    input_sum = x[0]+x[1]
    #print(input_sum)
    if input_sum<2:
      Output =1
    else:
      Output=0

  if gate == "NOR":

    ## Write your code here
    #Output = []
    input_sum = x[0]+x[1]
    #for i in range(input_sum[0]):
    if input_sum<1:
        Output=1
    else:
        Output=0
  
  if gate == "XOR":

    ## Write your code here
    #Output = []
    input_sum = x[0]+x[1]
    #for i in range(input_sum[0]):
    if input_sum==1:
        Output=1
    else:
        Output=0

  
  return Output

import itertools
inp_list = list(itertools.product([0,1],repeat=2))## Create a list consisting of inputs to logic gates (Binary, i.e (0,0),(0,1),(1,0),(1,1))
req_gates = ["NAND","NOR","XOR"]## Create a list of all gates (NAND,NOR,XOR)

for gate in req_gates:
  print('Results for ' + gate + ' gate : ')
  for inp in inp_list:
    inp = np.asarray(inp)
    #print(inp[1])
    print("Input is : " + str(inp))
    out = mp_model(inp,gate)
    print("Logic Gate output is : ",out)
  
  print('=====================================')
```

```python
## Perceptron method
#Step 1: generate a vector of inputs and a vector of weights
# matrix of inputs
input_table = np.array([
    [0,0], # both no
    [0,1], # one no, one yes
    [1,0], # one yes, one no
    [1,1]  # bot yes
])
# for NAND
print("=============================================")
print("for NAND gate")
print(f'input table:\n{input_table}')
# array of weights
weights = np.array([-1,-1])
print(f'weights: {weights}')

#Step 2: compute the dot product between the matrix of inputs and weights
# dot product matrix of inputs and weights
dot_products = input_table @ weights
print(f'Dot products: {dot_products}')

#Step 3: define the threshold activation function
def linear_threshold_gate(dot: int, T: float) -> int:
    '''Returns the binary threshold output'''
    if dot >= T:
        return 1
    else:
        return 0
T = -1
for i in range(0,4):
    activation = linear_threshold_gate(dot_products[i], T)
    print(f'Activation: {activation}')
```

```python
# for NOR
print("=============================================")
print("for NOR gate")
print(f'input table:\n{input_table}')
# array of weights
weights = np.array([-1,-1])
print(f'weights: {weights}')

#Step 2: compute the dot product between the matrix of inputs and weights
# dot product matrix of inputs and weights
dot_products = input_table @ weights
print(f'Dot products: {dot_products}')

#Step 3: define the threshold activation function
def linear_threshold_gate(dot: int, T: float) -> int:
    '''Returns the binary threshold output'''
    if dot >= T:
        return 1
    else:
        return 0
T = 0
for i in range(0,4):
    activation = linear_threshold_gate(dot_products[i], T)
    print(f'Activation: {activation}')
```

```python
# for XOR
print("=============================================")
print("for XOR gate")
print(f'input table:\n{input_table}')
# array of weights
weights = np.array([-1,-1])
print(f'weights: {weights}')

#Step 2: compute the dot product between the matrix of inputs and weights
# dot product matrix of inputs and weights
dot_products = input_table @ weights
print(f'Dot products: {dot_products}')

#Step 3: define the threshold activation function
def linear_threshold_gate(dot: int, T: float) -> int:
    '''Returns the binary threshold output'''
    if dot == T:
        return 1
    else:
        return 0
T = -1
for i in range(0,4):
    activation = linear_threshold_gate(dot_products[i], T)
    print(f'Activation: {activation}')
```

```python
## Part B : Rosenblatt's Percept

import numpy as np
import matplotlib.pyplot as plt
#step function 
def step (net):
    if net >= 0:
        return 1
    else:
        return 0
def rosenblatt_perceptron(inp_list,output,learning_rate,w_initial,num_epochs,gate):

  error = []
  
  #delta_w =0
  for epoch in range(num_epochs):
    out = [0,0,0,0]
    w_1 = w_initial[0]
    w_2 = w_initial[1]
    theta = w_initial[2]
    err = np.array([0,0,0,0])
    ## Write the learning code here, return new weights and error, Save the error value after each iteration to plot a error v/s iteration graph in case of all 3 gates
    for i in range(len(inp_list)):
      f_net = step(inp_list[i].dot(([w_1, w_2]))  + theta)
      err[i] = output[i] - f_net
      w_1 = w_1 + learning_rate * err[i] * inp_list[i][0]
      w_2 = w_2 + learning_rate * err[i] * inp_list[i][1]
      theta = theta + learning_rate*err[i]
      out[i] = f_net
    final_err = np.sum(err)
    # gradient based perceptron learning law for weight updation
    #delta_w=(learning_rate*(output-output_value).T.dot(input)).reshape(3,1)
    #w_new=w_initial+delta_w
    w_new = np.concatenate((w_1.reshape(1,1),w_2.reshape(1,1),theta.reshape(1,1)))
    w_initial = w_new
    #err = output - output_value
    error.append(final_err)

  return w_new,error, out
#input with bias as third column
inp = [[0,0],[0,1],[1,0],[1,1]]
inp = np.asarray(inp)
i0=np.ones(input.shape[0]).reshape(input.shape[0],1)
input_i0=np.concatenate((inp,i0),axis=1)
inp_list = inp## Create a list consisting of inputs to logic gates (Binary, i.e (0,0),(0,1),(1,0),(1,1))
nand_output = np.array([1,1,1,0]).reshape(4,1)## Define NAND gate target output
nor_output = np.array([1,0,0,0]).reshape(4,1)## Define NOR gate target output
xor_output = np.array([0,1,1,0]).reshape(4,1)## Define XOR gate target output

learning_rate = 0.1## Define a learning rate
w_initial = np.random.uniform(0,1,(input_i0.shape[1],1))## Initialise weights
num_epochs = 50 ## Set number of epochs

## For NAND gate
print("=========================================")
print("For NAND gate")
w_new,error, nand_out = rosenblatt_perceptron(inp_list,nand_output,learning_rate,w_initial,num_epochs,"NAND")
print(w_new)
print(nand_out)
plt.plot(error)
plt.show()
## For NOR gate
print("=========================================")
print("For NOR gate")
w_new,error, nor_out = rosenblatt_perceptron(inp_list,nor_output,learning_rate,w_initial,num_epochs,"NOR")
print(w_new)
print(nor_out)
plt.plot(error)
plt.show()
## For XOR gate
print("=========================================")
print("For XOR gate")
w_new,error, xor_out = rosenblatt_perceptron(inp_list,xor_output,learning_rate,w_initial,num_epochs,"XOR")
print(w_new)
print(xor_out)
plt.plot(error)
plt.show()
```

## Inferences and Conclusion : State all the key observations and conclusion

**Observation**

The output for NAND Gate, NOR Gate are accurate and linearly separable while for XOR Gate is not linearly separable. 

**For MP Neuron-**

1. We have to find threshold value manually/brute force method.

2. All the inputs have equal importance.

3. Threshold 'b' can take only few possible values.

4. Unable to classify non-linear separable data.

**For perceptron-**

1. Learn weights to classify linear separable data.

2. Single perceptron is unable to classify Non-linear separable data. 

3. Whether we will converge to correct weight values will depend on our initialization of learning rate. Very small learning rate(of order 0.00001 or lesser) will give incorrect output.

4. Output values are boolean. Therefore, it cannot be used for regression task

```python
## Widrow Adaline model : Perform the experiment for different weight inialisations and learning rates and state your observations

def widrow_adaline(inp,learning_rate,w_initial,num_epochs,target):

  error = []
  for epoch in range(num_epochs):
    
    ## Write the learning(training) loop here and return the new weights, also save the error after each iteration
    weighted_sum = np.dot(inp,(w_initial))
    err = target - weighted_sum
    cost  = np.sum(np.square(err))/(inp).shape[0]
    error.append(cost)  
    #if cost!=0:
    w_new= w_initial + learning_rate*(inp).T.dot(err)    
    #else:
      #w_new = w_new  
    w_initial = w_new
  return w_new,error, weighted_sum  


N = 100 ## Set the number of input points 
learning_rate = 0.0000001 ## Define a learning rate
w_initial = np.array([[1],[1],[1]]) ## Initialise weights
num_epochs = 500 ## Set number of epochs

## Generate the input below 

#generating true output values
arr=np.array([2,2,1]).reshape(3,1)
y_target=((x).dot(arr))
inp = x ## An input array of the shape (3,100)

w_new,error, out = widrow_adaline(inp,learning_rate,w_initial,num_epochs, y_target)
print("Final weights for learning rate " + str(learning_rate) +" are:-", w_new)
plt.plot(error, marker ='*')
plt.show()
#w_new
```
